## 4. Compute tf-idf

```python
def compute_tf_idf(encoded_string):
    total = len(encoded_string)
    counter =  Counter(encoded_string)
    tf = {k: v / total for k, v in counter.items()}
    idf = 1 # as soon as we have only one string (doc)
    
    tf_idf = {k: v * idf for k, v in tf.items()}
    return tf_idf
```

```python
def show_plot(data):
    plt.figure(figsize=(10, 10))
    letters, rate = zip(*sorted(data.items(), key=lambda x: x[0]))
    plt.bar(letters, np.array(rate))
    plt.grid()
```

```python
def analyze(data, map_dict, verbose=1):
    substituted = decrypt_cypher(data, map_dict)
    res = compute_tf_idf(substituted)
    if verbose:
        show_plot(res)
    return substituted
```

```python
def generate_map_dict():
    tmp_letters = list(LETTERS)
    np.random.shuffle(tmp_letters)
    return dict(zip(LETTERS, tmp_letters))
```

```python
def update_map_dict(map_dict, key, value):
    value_key = [k for k, v in map_dict.items() if v == value][0]
    map_dict[key], map_dict[value_key] = map_dict[value_key], map_dict[key]
    
    return map_dict
```

```python
def generate_population(inputs, n_top_symbols=4):
    top_letters = 'etaoinshrdlu'[:n_top_symbols]
    top_letters_perm = [''.join(p) for p in permutations(top_letters)]
    top_letters_in_s = ''.join([x[0] for x in sorted(Counter(inputs).items(), key=lambda x: x[1], reverse=True)])
    
    population = [
        generate_map_dict()
        for _ in range(len(top_letters_perm))
    ]
    
    for i in range(len(population)):
        for c1, c2 in zip(top_letters_in_s, top_letters_perm[i]):
            update_map_dict(population[i], c1, c2)
    return population
```

```python
def compute_fitness(inputs, n_gram_size=N_GRAM_SIZES):
    global MOST_COMMON_WORDS
    
    score = 0
    for gram_size in n_gram_size:
        for index in range(len(inputs) - gram_size):
            word = inputs[index:index+gram_size]
            if word in MOST_COMMON_WORDS:
                score += 1
    return score
```

```python
def mating_pool(inputs, populations, num_parents):
    max_scores = [-999 * num_parents]
    parents = []
    for pop in populations:
        res = analyze(inputs, pop, verbose=0)
        current_score = compute_fitness(res)
        
        max_scores.append(current_score)
        parents.append(pop)
        
    tmp = np.asarray(max_scores)
    max_index = tmp.argsort()[num_parents:]
    return sorted(dict(zip(max_scores, parents)).items(), key=lambda x: x[0], reverse=True)[:num_parents]
```

```python
def crossover(populations, parents):
    offsprings = []
    offspring_size = len(populations) - len(parents)
    for k in range(offspring_size):
        offspring = {}
        parent_index1 = k % len(parents)
        parent_index2 = (k + 1) % len(parents)
        
        counter = 0
        for k, v in parents[parent_index1].items():
            if counter <= len(parents[parent_index1]) // 2:
                offspring[k] = v
            counter -= 1

        counter = 0
        for k, v in parents[parent_index1].items():
            if counter > len(parents[parent_index2]) // 2:
                offspring[k] = v
            counter += 1

        offsprings.append(offspring)

    return offsprings
```

```python
def mutate(dicts, mutation_proba=0.25):
    for i in range(len(dicts)):
        while np.random.rand() < mutation_proba:
            idx1 = np.random.choice(list(LETTERS))
            idx2 = np.random.choice(list(LETTERS))
            
            update_map_dict(dicts[i], idx1, idx2)
                

    return dicts
```

```python
def algorithm(inputs, num_generations):
    print('Start genetic algorithm...')
    
    max_score = -999
    best_dict = None
    new_population = generate_population(inputs)
    for i in range(num_generations):
        print(f'Generation {i}')
        parents_with_scores = mating_pool(inputs, new_population, len(new_population))
        scores, parents = parents_with_scores.keys(), parents_with_scores.values()
        tmp = np.asarray(list(scores))
        max_index = tmp.argmax()
        if tmp[max_index] > max_score:
            max_score = tmp[max_index]
            best_dict = list(parents)[max_index]
            print(f'Current best score {max_score}')

        offsprings = crossover(new_population, list(parents))
        
        _population = []
        for offspring in offsprings:
            _population.append(offspring)
        for parent in parents:
            _population.append(parents)
            
        new_population = _population
        print(f'Population size: {len(new_population)}')
    print(f'Best score: {max_score}')
    print(f'Best dict: {best_dict}')
    if best_dict != None:
        print(f'Result: {decrypt_cypher(inputs, best_dict)}')
```
